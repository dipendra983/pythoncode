```python
class Vehicles:
  def __init__(self,make,model,year):
    self.make = make
    self.model = model
    self.year = year

  def start_engine(self):
    print(f" make {self.make} model {self.model} year {self.year}")

  def stop_engine(self):
    print(f" make {self.make} model {self.model} year {self.year}")
```

```python
class car(Vehicles):
  def __init__(self,make,model,year):
   super().__init__(make,model,year)

  def start_engine(self):
    print(f"car {self.make} {self.model} {self.year}")

  def stop_engine(self):
    print(f"car {self.make} {self.model} {self.year}")
```

```python
class motorcycle(Vehicles):
  def __init__(self,make,model,year):
   super().__init__(make,model,year)

  def start_engine(self):
    print(f"motorcycle {self.make} {self.model} {self.year}")

  def stop_engine(self):
    print(f"motorcycle {self.make} {self.model} {self.year}")
```

```python
car = car("toyato","fortuner",2024)
car.start_engine()
car.stop_engine()
```

```python
vehicles = Vehicles("toyato","fortuner","2024")
vehicles.start_engine()
vehicles.stop_engine()
```

```python
motorcycle = motorcycle("royal enfield","classic 350",2023)
motorcycle.start_engine()
motorcycle.stop_engine()
```

```python
class LibraryItem:
  def __init__(self,title,author,publication_year):
    self.title = title
    self.author = author
    self.publication_year = publication_year

  def checkout(self):
    print(f"Title: {self.title}\nAuthor: {self.author}\nYear: {self.publication_year}")


  def return_item(self):
    print(f"Title: {self.title}\nAuthor: {self.author}\nYear: {self.publication_year}")
```

```python
class book(LibraryItem):
  def __init__(self,title,author,publication_year):
    super().__init__(title,author,publication_year)

  def checkout(self):
    print(f"Title: {self.title}\nAuthor {self.author}\nYear {self.publication_year}")


  def return_item(self):
    print(f"Title: {self.title}\nAuthor: {self.author}\nYear: {self.publication_year}")
```

```python
class magazine(LibraryItem):
  def __init__(self,title,author,publication_year):
    super().__init__(title,author,publication_year)

  def checkout(self):
    print(f"Title: {self.title}\nAuthor: {self.author}\nYear: {self.publication_year}")


  def return_item(self):
    print(f"Title: {self.title}\nAuthor: {self.author}\nYear: {self.publication_year}")
```

```python
book = book("the adventures of tom swyer","mark twain","2006")
book.checkout()
book.return_item()
```

```python
magazine = magazine("the adventures of tom swyer","mark twain","2001")
magazine.checkout()
magazine.return_item()
```

```python
class Atm:
  def __init__(self):
    self.pin=""
    self.balance=0

    self.menu()

  def menu(self):
    user_input=input("""
    Hi how can I help you?
    1. enter to create a pin
    2. enter to deposit
    3. enter to withdraw
    4. enter to check balance
    5. enter to exit
    """)
    if user_input=="1":
      self.create_pin()
    elif user_input=="2":
      self.deposit()
    elif user_input=="3":
      self.withdraw()
    elif user_input=="4":
      self.check_balance()
    else:
      exit()
```
